Visualizing Using a chart

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(13, 4))
sns.countplot(x='LateSubmission', data=Reporting)
plt.title('Reporting Analysis')
plt.xlabel('LateSubmission')
plt.ylabel('ReportErrors')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(13, 4))
sns.scatterplot(x='LateSubmission', y='ReportErrors', data=Reporting)
plt.title('Reporting Analysis')
plt.xlabel('LateSubmission')
plt.ylabel('ReportErrors')
plt.show()
```
